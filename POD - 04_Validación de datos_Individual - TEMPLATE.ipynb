Hemos detectado que ciertos clientes tienen nacionalidades que no existen. Ademas, hay numeros de documento negativos y se tiene que poder agregar la fecha de registro de los clientes (que no pueden ser del futuro). Cambiar la definicion del cliente para que estas cosas no ocurran. Despues de la siguiente celda, cree otras 3 mas probando un caso donde deberia funcionar y otros dos en los que no.

````python
#Escribir el codigo para responder la pregunta en esta celda
from datetime import datetime, date
from typing import List

class Client:
    # Esta lista me la re invente porque no se como hacerlo de otra forma
    NACIONALIDADES_VALIDAS: List[str] = [
        "Española",
        "Argentina",
        "Colombiana",
        "Peruana",
        "Chilena",
        "Mexicana",
        "Uruguaya",
        "Venezolana",
        "Boliviana",
        "Paraguaya"
    ]

    def __init__(self, dni: int, nacionalidad: str, fecha_registro: str):
        self.dni = dni
        self.nacionalidad = nacionalidad
        self.fecha_registro = fecha_registro
        self._validar_dni()
        self._validar_nacionalidad()
        self._validar_fecha_registro()

    def _validar_dni(self):
        if not isinstance(self.dni, int):
            raise TypeError(f"El DNI debe ser un entero.")
        if self.dni <= 0:
            raise ValueError("El DNI debe ser mayor o igual a 0.")

    def _validar_nacionalidad(self):
        if not isinstance(self.nacionalidad, str):
            raise TypeError(f"La nacionalidad debe ser una string.")
        if self.nacionalidad not in self.NACIONALIDADES_VALIDAS:
            raise ValueError(f"Nacionalidad '{self.nacionalidad}' no es válida.")

    def _validar_fecha_registro(self):
        if not isinstance(self.fecha_registro, str):
            raise TypeError(f"La fecha de registro tiene que ser YYYY-MM-DD.")
        try:
            fecha = datetime.strptime(self.fecha_registro, "%Y-%m-%d").date()
        except ValueError:
            raise ValueError("La fecha de registro tiene que tener el formato YYYY-MM-DD.")
        if fecha > date.today():
            raise ValueError("La fecha de registro no puede ser del futuro.")
        self.fecha_registro = fecha

    def __repr__(self):
        return f"Client(dni={self.dni}, nacionalidad='{self.nacionalidad}', fecha_registro={self.fecha_registro})"
````

````python
#Escribir el codigo para responder la pregunta en esta celda
try:
    cliente_valido = Client(
        dni=12345678,
        nacionalidad="Española",
        fecha_registro="2023-05-15"
    )
    print("Cliente válido creado exitosamente:")
    print(cliente_valido)
except (TypeError, ValueError) as e:
    print(f"Error al crear el cliente válido: {e}")
````

````python
#Escribir el codigo para responder la pregunta en esta celda
try:
    cliente_invalido_nacionalidad = Client(
        dni=87654321,
        nacionalidad="ASAFAEFE",
        fecha_registro="2023-06-20"
    )
except (TypeError, ValueError) as e:
    print(f"Error al crear el cliente con nacionalidad inválida: {e}")
````
